###Data Visualizations with Matplotlib and Seaborn (Python)

Educational program with **3 practical examples** of charts to analyze data using **Pandas, Matplotlib, and Seaborn**. Ideal for learning data visualization in a simple and effective way.

**What exactly this code does**

#### Exercise 1 – Bar Chart
- Creates a DataFrame with sales of 8 products in 3 cities (Bogotá, Cali, Medellín)
- Generates a **bar chart** comparing sales by product and city
- Uses Seaborn for a nice style and Matplotlib to display the chart

```python
# Bar Chart
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid", palette="Set2")

products = ['Product A', 'Product B', 'Product C',
             'Product D', 'Product E', 'Product F',
             'Product G', 'Product H']

sales = {
    'Bogota': [120, 130, 110, 100, 90, 100, 105, 90],
    'Cali': [100, 120, 130, 140, 150, 160, 155, 160],
    'Medellin': [90, 100, 150, 200, 250, 300, 400, 80]
}

df_sales = pd.DataFrame(sales, index=products)
print(df_sales)

plt.figure(figsize=(5, 5))
df_sales.plot(kind='bar')
plt.title("Product Sales by City")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

#### Exercise 2 – Line Chart
- Creates a DataFrame with monthly sales (8 months) of 3 products in different cities
- Generates a **line chart** with markers to visualize monthly trends
- Shows how sales evolve over time

```python
# Grafica de lineas
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid", palette="Set2")

months = ['January', 'February', 'March',
             'April', 'May', 'June',
             'July', 'August']
df_products = pd.DataFrame({
    'Month': months,
    'Potato (Bogotá)': [120, 130, 110, 100, 90, 100, 105, 90],
    'Onion (Cali)': [100, 120, 130, 140, 150, 160, 155, 160],
    'Tomato (Medellín)': [90, 100, 150, 200, 250, 300, 400, 80]
})

plt.figure(figsize=(10, 6))
df_products.set_index("Month").plot(marker='o')
plt.title("Monthly product sales by city")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()

```

#### Exercise 3 – Scatter Plot
- Creates a DataFrame with study hours, score, and motivation level
- Generates a **scatter plot** where:
  - The size and color of the points represent the motivation level
  - Includes a color bar to interpret the motivation

```python
# Scatter Plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df_estudio = pd.DataFrame({
    'Hours': list(range(0, 21)),
    'Score' : [50, 60, 65, 70, 75, 78, 80, 82, 85, 87,
                    90, 92, 94, 96, 99, 100, 98, 80, 87, 89, 100],
    'Motivacion' : [20, 25, 30, 35, 40, 45, 50, 55, 60, 65,
                    70, 75, 80, 85, 100, 98, 94, 99, 95, 89, 90]
})

plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_estudio['Hours'], df_estudio['Score'],
                      s=df_estudio['Motivacion'], c=df_estudio['Motivacion'],
                      cmap='viridis', alpha=0.7)
plt.title("Relationship between study hours and motivation")
plt.xlabel("Study hours")
plt.ylabel("Score")
plt.colorbar(scatter, label="Motivation level")
plt.show()
```
